```python
# Importing modules
import csv
from xlsxwriter.workbook import Workbook

import pandas as pd
  
```

```python
# title.akas.tsv.gz to get movie_id, movie_name, country

# Read TSV file into DataFrame
df1 = pd.read_table('title.akas.tsv.gz')
#print(df)

# Error: Your sheet size is: 33475127, 8 Max sheet size is: 1048576, 16384
# drop rows but still get 30 million rows
df1 = df1[0:30000000] 
print(df1)

# drop rows with null values
df1.dropna()

# rename column titles to match attribute names --> old:new
df1.rename(columns = {'titleId':'movie_id', 'title':'movie_name', 'region':'country'}, inplace = True)


# delete rows with movie_id duplicates and keep US version
duplicates = df1.duplicated(subset=['movie_id'])
if duplicates.any():
    deleteRow = df1[df1['country'] != 'US'].index
    df1.drop(deleteRow, inplace=True)
    # if duplicates in movie_id still, then keep the first one 
    df1 = df1.sort_values('ordering', ascending=False).drop_duplicates('movie_id').sort_index()
    
# columns to get rid of from this sheet:
del df1['ordering']
del df1['language']
del df1['types']
del df1['attributes']
del df1['isOriginalTitle']

print(df1)

# create excel writer object
#writer = pd.ExcelWriter('movies_test1.xlsx') 
# write dataframe to excel
#df1.to_excel(writer)
# save the excel
#writer.save()
print('DataFrame is written successfully to Excel File.')
```

```python
# title.basics.tsv.gz for release_year, genre, length 

# Read TSV file into DataFrame
df4 = pd.read_table('title.basics.tsv.gz')
#print(df2)

# limit df size (number of rows printing)
# Your sheet size is: 9297065, 4 Max sheet size is: 1048576, 16384
df4 = df4[0:1048576] 
print(df4)

# drop rows with null values
df4.dropna()

# rename column titles to match attribute names --> old:new
df4.rename(columns = {'tconst':'movie_id', 'startYear':'release_year', 'runtimeMinutes':'length', 'genres':'genre'}, inplace = True)

# drop rows with titleType not as movie 
df4.drop(df4[df4['titleType'] != 'movie'].index, inplace=True)
#df4.drop(df4[df4['titleType'] != 'tvMovie'].index, inplace=True)
    
# columns to get rid of from this sheet: 
del df4['titleType']
del df4['primaryTitle']
del df4['originalTitle']
del df4['isAdult']
del df4['endYear']

# create excel writer object
writer = pd.ExcelWriter('movies_test4.xlsx') 
# write dataframe to excel
df4.to_excel(writer)
# save the excel
writer.save()
print('DataFrame is written successfully to Excel File.')
```
